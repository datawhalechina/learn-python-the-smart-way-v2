```python
# æˆ‘ä¸çŸ¥é“å®ƒä»€ä¹ˆæ—¶å€™åœä¸‹æ¥

def leftmostDigit(n):
    n = abs(n)
    while n >= 10:
        n = n//10
    return n 
```

ä¸¾ä¸ªä¾‹å­ï¼šä¾æ¬¡æ‰¾å‡º n ä¸ª 4 æˆ–è€… 7 çš„æ•´æ•°å€éè´Ÿæ•´æ•°

```python
def isMultipleOf4or7(x):
    return ((x % 4) == 0) or ((x % 7) == 0)

def nthMultipleOf4or7(n):
    found = 0
    guess = -1
    while found <= n:
        guess += 1
        if isMultipleOf4or7(guess):
            found += 1
    return guess
```

```python
print("Multiples of 4 or 7: ", end="")
for n in range(15):
    print(nthMultipleOf4or7(n), end=" ")
```

<big>isPrime</big>

åˆ¤æ–­ä¸€ä¸ªæ•°æ˜¯ä¸æ˜¯è´¨æ•°

```python
# ä¸æ˜¯æœ€å¿«çš„å†™æ³•ï¼Œä½†æœ€å®¹æ˜“ç†è§£

def isPrime(n):
    if n < 2:
        return False
    for factor in range(2,n):
        if n % factor == 0:
            return False
    return True
```

```python
for n in range(100):
    if isPrime(n):
        print(n, end=" ")
```

<big>faster IsPrime</big>:


```python
# å¿«äº†ä¸€ç‚¹

def fasterIsPrime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    maxFactor = round(n**0.5)
    for factor in range(3, maxFactor+1, 2):
        if n % factor == 0:
            return False
    return True
```

```python
for n in range(100):
    if fasterIsPrime(n):
        print(n, end=" ")
```

```python
# éªŒè¯ä»–å®ƒä¿©ç»“æœæ˜¯ä¸€æ ·çš„
for n in range(100):
    assert(isPrime(n) == fasterIsPrime(n))
print("They seem to work the same!")
```

<big>nth Prime</big>

ä¾æ¬¡æ‰¾å‡ºç¬¬ n ä½è´¨æ•°

```python
def nthPrime(n):
    found = 0
    guess = 0
    while found <= n:
        guess += 1
        if fasterIsPrime(guess):
            found += 1
    return guess
```

```python
for n in range(10):
    print(n, nthPrime(n))
print("Done!")
```
