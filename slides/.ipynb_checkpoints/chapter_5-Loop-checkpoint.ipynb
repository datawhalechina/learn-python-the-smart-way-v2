```python
def sumEveryKthFromMToN(m, n, k):
    total = 0
    # ç¬¬ä¸‰ä¸ªå‚æ•°ä¸º â€œæ­¥é•¿â€ step
    for x in range(m, n+1, k):
        total += x
    return total
```

åªå¯¹ä» `m` åˆ° `n` çš„ **å¥‡æ•°æ±‚å’Œ**

```python
# æˆ‘ä»¬ä¹Ÿå¯ä»¥é€šè¿‡ä¿®æ”¹å¾ªç¯å†…éƒ¨çš„ä»£ç æ¥æ”¹å˜æ­¥é•¿

def sumOfOddsFromMToN(m, n):
    total = 0
    for x in range(m, n+1):
        if x % 2 == 1:
            total += x
    return total
```

```python
sumOfOddsFromMToN(4, 10) == sumOfOddsFromMToN(5,9) == (5+7+9)
```

ç°åœ¨æˆ‘ä»¬**åç€æ¥**è¯•ä¸€ä¸‹ï¼

```python
# æˆ‘ä»¬å°†ç”Ÿæˆä¸€ä¸ªåå‘æ•°å­—åºåˆ—
# (ä»…ä¾›æ¼”ç¤ºä½¿ç”¨ï¼Œä»£ç å®è·µä¸­ä¸å»ºè®®è¿™ä¹ˆåš)

def sumOfOddsFromMToN(m, n):
    total = 0
    for x in range(n, m-1, -1):
        if x % 2 == 1:
            total += x
    return total
```

```python
sumOfOddsFromMToN(4, 10) == sumOfOddsFromMToN(5,9) == (5+7+9)
```

```python
# ä¸‹é¢çš„ä»£ç å°†è¾“å‡ºäºŒç»´åæ ‡

def printCoordinates(xMax, yMax):
    for x in range(1, xMax+1):
        for y in range(1, yMax+1):
            print(f"( {x} , {y} )  ", end="")
        print()
```

# while å¾ªç¯

<big>å˜¿ï¼</big>

å½“ä½ **ä¸çŸ¥é“å¾ªç¯ä»€ä¹ˆæ—¶å€™åœä¸‹æ¥**çš„æ—¶å€™ï¼Œä¸ºä»€ä¹ˆä¸è¯•è¯• **`while`**

```python
# æˆ‘ä¸çŸ¥é“å®ƒä»€ä¹ˆæ—¶å€™åœä¸‹æ¥

def leftmostDigit(n):
    n = abs(n)
    while n >= 10:
        n = n//10
    return n
```

```python
leftmostDigit(72658489290098) == 7
```

ä¸¾ä¸ªä¾‹å­ï¼šä¾æ¬¡æ‰¾å‡º n ä¸ª 4 æˆ–è€… 7 çš„æ•´æ•°å€éè´Ÿæ•´æ•°

```python
def isMultipleOf4or7(x):
    return ((x % 4) == 0) or ((x % 7) == 0)

def nthMultipleOf4or7(n):
    found = 0
    guess = -1
    while found <= n:
        guess += 1
        if isMultipleOf4or7(guess):
            found += 1
    return guess
```

```python
print("Multiples of 4 or 7: ", end="")
for n in range(15):
    print(nthMultipleOf4or7(n), end=" ")
```

```python
# ä¸éœ€è¦çœ‹æ‡‚è¿™äº›ä»£ç ï¼Œå…³æ³¨æ¼”ç¤ºçš„è¿‡ç¨‹

def readUntilDone():
    linesEntered = 0
    while True:
        response = input("Enter a string (or 'done' to quit): ")
        if response == "done":
            break
        print("  You entered: ", response)
        linesEntered += 1
    print("Bye!")
    return linesEntered

linesEntered = readUntilDone()
print("You entered", linesEntered, "lines (not counting 'done').")
```

# isPrime

åˆ¤æ–­ä¸€ä¸ªæ•°æ˜¯ä¸æ˜¯è´¨æ•°

```python
# ä¸æ˜¯æœ€å¿«çš„å†™æ³•ï¼Œä½†æœ€å®¹æ˜“ç†è§£

def isPrime(n):
    if n < 2:
        return False
    for factor in range(2,n):
        if n % factor == 0:
            return False
    return True
```

```python
for n in range(100):
    if isPrime(n):
        print(n, end=" ")
```

## faster IsPrime:


```python
# å¿«äº†ä¸€ç‚¹

def fasterIsPrime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    maxFactor = round(n**0.5)
    for factor in range(3,maxFactor+1,2):
        if n % factor == 0:
            return False
    return True
```

```python
for n in range(100):
    if fasterIsPrime(n):
        print(n, end=" ")
```

## çœŸçš„å¿«äº†ï¼Ÿ

```python
# éªŒè¯ä»–å®ƒä¿©ç»“æœæ˜¯ä¸€æ ·çš„
for n in range(100):
    assert(isPrime(n) == fasterIsPrime(n))
print("They seem to work the same!")
```

```python
import time

bigPrime = 102030407
print("Timing isPrime(",bigPrime,")", end=" ")

# isPrime
time0 = time.time()
print(", returns ", isPrime(bigPrime), end=" ")

time1 = time.time()
print(", time = ",(time1-time0)*1000,"ms\n")

# fasterIsPrime
print("Timing fasterIsPrime(",bigPrime,")", end=" ")
time0 = time.time()

print(", returns ", fasterIsPrime(bigPrime), end=" ")
time1 = time.time()

# result
print(", time = ",(time1-time0)*1000,"ms")
```

## nthPrime

ä¾æ¬¡æ‰¾å‡ºç¬¬ n ä½è´¨æ•°

```python
def nthPrime(n):
    found = 0
    guess = 0
    while found <= n:
        guess += 1
        if fasterIsPrime(guess):
            found += 1
    return guess
```

```python
for n in range(10):
    print(n, nthPrime(n))
print("Done!")
```
