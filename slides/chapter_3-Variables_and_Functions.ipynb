# Helper Functions

ç¼–å†™å‡½æ•°æ˜¯ç”¨æ¥è§£å†³é—®é¢˜çš„

è¿™ç§å‡½æ•°å°±å«åš `Helper Function`

```python
def onesDigit(n):
    return n%10

def largerOnesDigit(x, y):
    return max(onesDigit(x), onesDigit(y))

print(largerOnesDigit(134, 672)) # 4
print(largerOnesDigit(132, 674)) # ä¾ç„¶æ˜¯ 4
```
